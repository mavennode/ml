# 猫的训练集图片宽高散点分布

```python
from PIL import Image

import os,re

# 列出data/train/cats中所有图片
def list_pictures(directory, ext='jpg|jpeg|bmp|png|ppm'):
    return [os.path.join(root, f)
            for root, _, files in os.walk(directory) for f in files
            if re.search(r'([\w]+\.(?:' + ext + '))', f)]

files = list_pictures('data/train/cats', 'jpg')

# x图片宽，y图片高
x = []
y = []

for file in files:
    im = Image.open(file)#返回一个Image对象
#     print('宽：%d,高：%d'%(im.size[0],im.size[1]))
    x.append(im.size[0])
    y.append(im.size[1])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()  
ax1 = fig.add_subplot(111)  
#设置标题  
ax1.set_title('cat-wide-height') 
# print(x)
# print(y)
plt.xlabel("wide")
plt.ylabel("height")
plt.axis([0, 600, 0, 600])
plt.scatter(x, y)
plt.show()
```
