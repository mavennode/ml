# jieba分词

```python
import jieba
import jieba.analyse
#导入自定义词典
# jieba.load_userdict("dict.txt")

text = '城市绿化是栽种植物以改善城市环境的活动。 城市绿化作为城市生态系统中的还原组织 城市生态系统具有受到外来干扰和破坏而恢复原状的能力，就是通常所说的城市生态系统的还原功能。'
#精确模式
seg_result = jieba.cut(text, cut_all=False)
print(u"分词结果:")
print("/".join(seg_result))

#获取关键词
tags = jieba.analyse.extract_tags(text, topK=3)
print(u"关键词:")
print(" ".join(tags))
```

## 读取各文件并分词

```python
import sys
import re
import codecs
import os
import shutil
import jieba
import jieba.analyse
#导入自定义词典
#jieba.load_userdict("dict.txt") 
# 停用词
stopwords = [line.strip() for line in open('文档分类/stopwords.txt', 'r', encoding='utf-8').readlines()]
for i in range(1,7):
    filename = '文档分类/00'+str(i)+'.txt'
    resname = '文档分类/00'+str(i)+'_result.txt'
    source = open(filename, 'r') 
    result = codecs.open(resname, 'w', 'utf-8')
    line = source.readline() 
    line = line.rstrip('\n')
    while line!="":
        seglist = jieba.cut(line,cut_all=False)  #精确模式
        outstr=''
        for word in seglist:
            if word not in stopwords:
                if word != '\t':
                    outstr += word
                    outstr += ' '
#         output = ' '.join(list(seglist))         #空格拼接
        print(outstr)
        result.write(outstr + '\r\n')
        line = source.readline() 
    else:
        print('End file: ' + str(i))
        source.close()
        result.close()
    i = i + 1
else:
    print("End All")
```

```python
result_file = '文档分类/result.txt'
result = codecs.open(result_file, 'w', 'utf-8')
j = 1
while j <= 6:
    fileName = '文档分类/00'+str(j)+'_result.txt'
    source = open(fileName, 'r')
    line = source.readline()
    line = line.strip('\n')
    line = line.strip('\r')
    
    while line!="":
        line = line.replace('\n',' ')
        line = line.replace('\r',' ')
        result.write(line+ ' ')
        line = source.readline()
    else:
        print('End file: ' + str(j))
        result.write('\r\n')
        source.close()
    j = j +1
    
else:
    print('End All')
    result.close()
```

# 计算TF-IDF

```python
# 文档语料，空格连接
corpus = []

# 读取语料，一行语料为一个文档
for line in open('文档分类/result.txt', 'r').readlines():
#     print(line)
    corpus.append(line.strip())
# print(corpus)

# 将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
X.toarray()
```

```python
# 获取词袋模型中的所有词语
word = vectorizer.get_feature_names()
print(word)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
# 统计每个词语的tf-idf权值，计算tf-idf
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(X)
weight = tfidf.toarray()
print(weight)
```

```python
res_tf_idf = '文档分类/tf_idf_result.txt'
result = codecs.open(res_tf_idf, 'w', 'utf-8')
for j in range(len(word)):
    result.write(word[j] + ' ')
result.write('\r\n\r\n')

# 第一个for遍历所有文本，第二个for遍历某一类文本下的词语权重
for i in range(len(weight)):
    for j in range(len(word)):
        result.write(str(weight[i][j]) + ' ')
    result.write('\r\n\r\n')
result.close()
```

# K-means聚类

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
model = clf.fit(weight)
# 3个中心点
print(clf.cluster_centers_)
```

```python
# 每个样本所属的簇
print(clf.labels_)
i = 1
while i <= len(clf.labels_):
    print(i, clf.labels_[i-1])
    i = i + 1
# 用来评估簇的个数是否合适，距离越小说明簇分的越好，选取临界点的簇个数
print(clf.inertia_)
```
